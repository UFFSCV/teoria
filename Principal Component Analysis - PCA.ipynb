## PCA + Visão Computacional

Partindo do entendimento do PCA, vemos que a redução das dimenções e que ele cria um agrupamento de caraterísticas, que são muito fáceis de entender em datasceinte mas como ele pode ser utilizado em visão computacional?

A resposta para isso está no __Eigenvector__ e no __Eigenvalues__

### Eigenvector  e Eigenvalues
$$ A\overrightarrow{\upsilon} = \lambda\overrightarrow{\upsilon}$$

$$ M = \left({\begin{array}{cc} 1 & 2 & 0 \\ 2 & 1 & 0 \\ 0 & 0 & -3 \end{array}}\right) X \left({\begin{array}{cc} 1 \\1 \\ 0  \end{array}}\right) = \left({\begin{array}{cc} 3 \\3 \\ 0  \end{array}}\right) = 3x \left({\begin{array}{cc} 1 \\1 \\ 0  \end{array}}\right)$$

$$ A\overrightarrow{\upsilon} = \left({\begin{array}{cc} 1 \\1 \\ 0  \end{array}}\right) $$
$$\lambda\overrightarrow{\upsilon} = 3 $$

Com essa transformação linear das imagens utilizando os Eigenvector e o Eigenvalue e fazemos a distânciação entre os pontos da imagem, pois a distância euclidiana não se comporta muito bem em espaços muito dimencionados, nos dando um conjunto de pequenas características que poderão ser utilizadados para identificação de padrões e assim poder fazer reconhecimentos e classificações através de uma IA, como uma CNN.
